### 1. Class, Object, Attribute, Method


```python
class Student:
    # Class attribute: shared by all objects
    university = "BRAC University"

    # Constructor: runs automatically when object is created
    def __init__(self, student_id, name, department, cgpa):
        self.student_id = student_id
        self.name = name
        self.department = department
        self.cgpa = cgpa

    # Instance method
    def show_info(self):
        print("Student ID:", self.student_id)
        print("Name:", self.name)
        print("Department:", self.department)
        print("CGPA:", self.cgpa)
        print("University:", self.university)

    def is_passed(self):
        if self.cgpa >= 2.00:
            return True
        return False


student1 = Student("CSE-101", "Smith", "CSE", 3.75)
student2 = Student("BBA-102", "Nila", "BBA", 2.90)

student1.show_info()
print("Passed:", student1.is_passed())

print("-" * 30)

student2.show_info()
print("Passed:", student2.is_passed())
```

### 2. Instance Variable vs Class Variable


```python
class Employee:
    company_name = "TechSoft Ltd"   # class variable
    employee_count = 0              # class variable

    def __init__(self, name, salary):
        self.name = name            # instance variable
        self.salary = salary        # instance variable
        Employee.employee_count += 1

    def show_employee(self):
        print(f"Name: {self.name}")
        print(f"Salary: {self.salary}")
        print(f"Company: {Employee.company_name}")
        print()


emp1 = Employee("Arafat", 30000)
emp2 = Employee("Mim", 45000)
emp3 = Employee("Hasan", 50000)

emp1.show_employee()
emp2.show_employee()
emp3.show_employee()

print("Total employees:", Employee.employee_count)

# Changing class variable
Employee.company_name = "NextGen Software"

emp1.show_employee()
emp2.show_employee()
```

### 3. Instance Method, Class Method, Static Method


```python
class Product:
    shop_name = "Smart Shop"
    total_products = 0

    def __init__(self, name, price, quantity):
        self.name = name
        self.price = price
        self.quantity = quantity
        Product.total_products += 1

    # Instance method: uses object data
    def total_price(self):
        return self.price * self.quantity

    # Class method: uses class data
    @classmethod
    def change_shop_name(cls, new_name):
        cls.shop_name = new_name

    @classmethod
    def show_total_products(cls):
        print("Total products created:", cls.total_products)

    # Static method: utility function
    @staticmethod
    def apply_discount(price, discount_percent):
        discount_amount = price * discount_percent / 100
        return price - discount_amount


p1 = Product("Laptop", 80000, 2)
p2 = Product("Mouse", 700, 5)

print(p1.name, "Total:", p1.total_price())
print(p2.name, "Total:", p2.total_price())

Product.show_total_products()

discounted_price = Product.apply_discount(1000, 10)
print("Discounted price:", discounted_price)

print("Old shop:", Product.shop_name)
Product.change_shop_name("Digital Mart")
print("New shop:", Product.shop_name)
```

### 4. Encapsulation with Private Variables


```python
class BankAccount:
    def __init__(self, account_no, holder_name, balance):
        self.account_no = account_no
        self.holder_name = holder_name
        self.__balance = balance  # private variable

    def deposit(self, amount):
        if amount <= 0:
            print("Deposit amount must be positive.")
            return

        self.__balance += amount
        print(f"{amount} deposited successfully.")

    def withdraw(self, amount):
        if amount <= 0:
            print("Withdraw amount must be positive.")
            return

        if amount > self.__balance:
            print("Insufficient balance.")
            return

        self.__balance -= amount
        print(f"{amount} withdrawn successfully.")

    def get_balance(self):
        return self.__balance

    def show_account(self):
        print("Account No:", self.account_no)
        print("Holder Name:", self.holder_name)
        print("Balance:", self.__balance)


account = BankAccount("AC-1001", "Sabbir", 10000)

account.show_account()
account.deposit(5000)
account.withdraw(3000)
print("Current balance:", account.get_balance())
```

### 5. Getter, Setter, and Property Decorator


```python
class StudentResult:
    def __init__(self, name, mark):
        self.name = name
        self.__mark = None
        self.mark = mark  # calls setter

    @property
    def mark(self):
        return self.__mark

    @mark.setter
    def mark(self, value):
        if 0 <= value <= 100:
            self.__mark = value
        else:
            raise ValueError("Mark must be between 0 and 100")

    @property
    def grade(self):
        if self.mark >= 98:
            return "A+"
        elif self.mark >= 90:
            return "A"
        elif self.mark >= 85:
            return "A-"
        elif self.mark >= 80:
            return "B+"
        elif self.mark >= 75:
            return "B"
        return "Fail"


result = StudentResult("Debashis", 85)
print(result.name)
print(result.mark)
print(result.grade)

result.mark = 95
print(result.mark)
print(result.grade)
```

### 10. Polymorphism


```python
class Rectangle:
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def area(self):
        return self.width * self.height


class Circle:
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return 3.1416 * self.radius * self.radius


class Triangle:
    def __init__(self, base, height):
        self.base = base
        self.height = height

    def area(self):
        return 0.5 * self.base * self.height


shapes = [
    Rectangle(10, 5),
    Circle(7),
    Triangle(10, 6)
]

for shape in shapes:
    print(type(shape).__name__, "Area:", shape.area())
```

### 11. Abstraction with ABC


```python
from abc import ABC, abstractmethod

class Vehicle(ABC):
    @abstractmethod
    def start_engine(self):
        pass

    @abstractmethod
    def stop_engine(self):
        pass


class Car(Vehicle):
    def start_engine(self):
        print("Car engine started")

    def stop_engine(self):
        print("Car engine stopped")


class Bike(Vehicle):
    def start_engine(self):
        print("Bike engine started")

    def stop_engine(self):
        print("Bike engine stopped")


vehicles = [Car(), Bike()]

for vehicle in vehicles:
    vehicle.start_engine()
    vehicle.stop_engine()
```

### 12. Magic Methods: `__str__`, `__repr__`, `__len__`, `__eq__`, `__add__`


```python
class Book:
    def __init__(self, title, author, pages, price):
        self.title = title
        self.author = author
        self.pages = pages
        self.price = price

    def __str__(self):
        return f"{self.title} by {self.author}"

    def __repr__(self):
        return f"Book(title='{self.title}', author='{self.author}', pages={self.pages}, price={self.price})"

    def __len__(self):
        return self.pages

    def __eq__(self, other):
        return self.title == other.title and self.author == other.author

    def __add__(self, other):
        return self.price + other.price


book1 = Book("Python OOP", "John Doe", 350, 500)
book2 = Book("Data Science", "Jane Smith", 420, 700)
book3 = Book("Python OOP", "John Doe", 350, 500)

print(book1)
print(repr(book1))
print("Pages:", len(book1))
print("Book1 == Book2:", book1 == book2)
print("Book1 == Book3:", book1 == book3)
print("Total price:", book1 + book2)
```

### 17. Project 1: Student Management System


```python
class Student:
    def __init__(self, student_id, name, department, cgpa):
        self.student_id = student_id
        self.name = name
        self.department = department
        self.cgpa = cgpa

    def update_cgpa(self, new_cgpa):
        if 0.0 <= new_cgpa <= 4.0:
            self.cgpa = new_cgpa
            print("CGPA updated successfully.")
        else:
            print("Invalid CGPA.")

    def get_grade_status(self):
        if self.cgpa >= 3.75:
            return "Excellent"
        elif self.cgpa >= 3.00:
            return "Good"
        elif self.cgpa >= 2.00:
            return "Passed"
        return "Failed"

    def show(self):
        print(f"ID: {self.student_id}")
        print(f"Name: {self.name}")
        print(f"Department: {self.department}")
        print(f"CGPA: {self.cgpa}")
        print(f"Status: {self.get_grade_status()}")
        print("-" * 30)


class StudentManagementSystem:
    def __init__(self):
        self.students = []

    def add_student(self, student_id, name, department, cgpa):
        if self.find_student(student_id):
            print("Student ID already exists.")
            return

        student = Student(student_id, name, department, cgpa)
        self.students.append(student)
        print("Student added successfully.")

    def find_student(self, student_id):
        for student in self.students:
            if student.student_id == student_id:
                return student
        return None

    def remove_student(self, student_id):
        student = self.find_student(student_id)
        if student:
            self.students.remove(student)
            print("Student removed successfully.")
        else:
            print("Student not found.")

    def update_student_cgpa(self, student_id, new_cgpa):
        student = self.find_student(student_id)
        if student:
            student.update_cgpa(new_cgpa)
        else:
            print("Student not found.")

    def show_all_students(self):
        if not self.students:
            print("No students found.")
            return

        for student in self.students:
            student.show()

    def show_department_students(self, department):
        found = False
        for student in self.students:
            if student.department.lower() == department.lower():
                student.show()
                found = True

        if not found:
            print("No student found in this department.")


sms = StudentManagementSystem()

sms.add_student("CSE-101", "Smith", "CSE", 3.80)
sms.add_student("CSE-102", "Nila", "CSE", 3.40)


print("\nCSE Students:")
sms.show_department_students("CSE")

print("\nRemove Student:")
sms.remove_student("CSE-101")

print("\nAfter Remove:")
sms.show_all_students()
```

### 18. Project 2: Bank Account System


```python
class BankAccount:
    next_account_number = 1001

    def __init__(self, holder_name, initial_balance=0):
        self.account_number = BankAccount.next_account_number
        BankAccount.next_account_number += 1

        self.holder_name = holder_name
        self.__balance = initial_balance
        self.transactions = []

        self.add_transaction("Account Created", initial_balance)

    def add_transaction(self, transaction_type, amount):
        self.transactions.append({
            "type": transaction_type,
            "amount": amount,
            "balance": self.__balance
        })

    def deposit(self, amount):
        if amount <= 0:
            print("Invalid deposit amount.")
            return False

        self.__balance += amount
        self.add_transaction("Deposit", amount)
        print(f"{amount} deposited successfully.")
        return True

    def withdraw(self, amount):
        if amount <= 0:
            print("Invalid withdraw amount.")
            return False

        if amount > self.__balance:
            print("Insufficient balance.")
            return False

        self.__balance -= amount
        self.add_transaction("Withdraw", amount)
        print(f"{amount} withdrawn successfully.")
        return True

    def transfer(self, receiver_account, amount):
        if self.withdraw(amount):
            receiver_account.deposit(amount)
            self.add_transaction("Transfer Sent", amount)
            receiver_account.add_transaction("Transfer Received", amount)
            print("Transfer successful.")
            return True
        return False

    def get_balance(self):
        return self.__balance

    def show_account(self):
        print("Account Number:", self.account_number)
        print("Holder Name:", self.holder_name)
        print("Balance:", self.__balance)
        print("-" * 40)

    def show_transactions(self):
        print(f"Transaction history for {self.holder_name}")
        for transaction in self.transactions:
            print(transaction)
        print("-" * 40)


account1 = BankAccount("Smith", 10000)
account2 = BankAccount("Emma", 5000)

account1.show_account()
account2.show_account()

account1.deposit(2000)
account1.withdraw(1500)
account1.transfer(account2, 3000)

account1.show_account()
account2.show_account()

account1.show_transactions()
account2.show_transactions()
```

### 19. Project 3: Library Management System


```python
class Book:
    def __init__(self, book_id, title, author, copies):
        self.book_id = book_id
        self.title = title
        self.author = author
        self.copies = copies

    def is_available(self):
        return self.copies > 0

    def borrow_copy(self):
        if self.is_available():
            self.copies -= 1
            return True
        return False

    def return_copy(self):
        self.copies += 1

    def show(self):
        print(f"Book ID: {self.book_id}")
        print(f"Title: {self.title}")
        print(f"Author: {self.author}")
        print(f"Available Copies: {self.copies}")
        print("-" * 30)


class Member:
    def __init__(self, member_id, name):
        self.member_id = member_id
        self.name = name
        self.borrowed_books = []

    def borrow_book(self, book):
        if book.borrow_copy():
            self.borrowed_books.append(book)
            print(f"{self.name} borrowed '{book.title}'")
        else:
            print("Book not available.")

    def return_book(self, book_id):
        for book in self.borrowed_books:
            if book.book_id == book_id:
                book.return_copy()
                self.borrowed_books.remove(book)
                print(f"{self.name} returned '{book.title}'")
                return
        print("This book was not borrowed by this member.")

    def show_borrowed_books(self):
        print(f"Borrowed books by {self.name}:")
        if not self.borrowed_books:
            print("No borrowed books.")
        for book in self.borrowed_books:
            print("-", book.title)
        print("-" * 30)


class Library:
    def __init__(self):
        self.books = []
        self.members = []

    def add_book(self, book):
        self.books.append(book)
        print("Book added successfully.")

    def add_member(self, member):
        self.members.append(member)
        print("Member added successfully.")

    def find_book(self, book_id):
        for book in self.books:
            if book.book_id == book_id:
                return book
        return None

    def find_member(self, member_id):
        for member in self.members:
            if member.member_id == member_id:
                return member
        return None

    def borrow_book(self, member_id, book_id):
        member = self.find_member(member_id)
        book = self.find_book(book_id)

        if not member:
            print("Member not found.")
            return

        if not book:
            print("Book not found.")
            return

        member.borrow_book(book)

    def return_book(self, member_id, book_id):
        member = self.find_member(member_id)

        if not member:
            print("Member not found.")
            return

        member.return_book(book_id)

    def show_all_books(self):
        for book in self.books:
            book.show()


library = Library()

library.add_book(Book("B-101", "Python Crash Course", "Eric Matthes", 3))
library.add_book(Book("B-102", "Clean Code", "Robert C. Martin", 2))
library.add_book(Book("B-103", "Automate the Boring Stuff", "Al Sweigart", 1))

library.add_member(Member("M-01", "Nila"))
library.add_member(Member("M-02", "Emma"))

library.show_all_books()

library.borrow_book("M-01", "B-101")
library.borrow_book("M-01", "B-103")
library.borrow_book("M-02", "B-103")

member = library.find_member("M-01")
member.show_borrowed_books()

library.return_book("M-01", "B-103")

library.show_all_books()
```

### 20. Project 4: Shopping Cart System


```python
class Product:
    def __init__(self, product_id, name, price, stock):
        self.product_id = product_id
        self.name = name
        self.price = price
        self.stock = stock

    def reduce_stock(self, quantity):
        if quantity <= self.stock:
            self.stock -= quantity
            return True
        return False

    def increase_stock(self, quantity):
        self.stock += quantity

    def show(self):
        print(f"{self.product_id} | {self.name} | Price: {self.price} | Stock: {self.stock}")


class CartItem:
    def __init__(self, product, quantity):
        self.product = product
        self.quantity = quantity

    def subtotal(self):
        return self.product.price * self.quantity

    def show(self):
        print(f"{self.product.name} x {self.quantity} = {self.subtotal()}")


class ShoppingCart:
    def __init__(self):
        self.items = []

    def add_item(self, product, quantity):
        if quantity <= 0:
            print("Invalid quantity.")
            return

        if not product.reduce_stock(quantity):
            print("Not enough stock.")
            return

        for item in self.items:
            if item.product.product_id == product.product_id:
                item.quantity += quantity
                print("Cart updated.")
                return

        self.items.append(CartItem(product, quantity))
        print("Item added to cart.")

    def remove_item(self, product_id):
        for item in self.items:
            if item.product.product_id == product_id:
                item.product.increase_stock(item.quantity)
                self.items.remove(item)
                print("Item removed from cart.")
                return
        print("Item not found in cart.")

    def total(self):
        total_amount = 0
        for item in self.items:
            total_amount += item.subtotal()
        return total_amount

    def checkout(self):
        if not self.items:
            print("Cart is empty.")
            return

        print("Checkout successful.")
        print("Total bill:", self.total())
        self.items.clear()

    def show_cart(self):
        print("Cart Items:")
        if not self.items:
            print("Cart is empty.")
        for item in self.items:
            item.show()
        print("Total:", self.total())
        print("-" * 40)


products = [
    Product("P-01", "Keyboard", 1200, 10),
    Product("P-02", "Mouse", 600, 20),
    Product("P-03", "Monitor", 15000, 5)
]

cart = ShoppingCart()

print("Available Products:")
for product in products:
    product.show()

cart.add_item(products[0], 2)
cart.add_item(products[1], 3)
cart.add_item(products[2], 1)

cart.show_cart()

cart.remove_item("P-02")
cart.show_cart()

cart.checkout()

print("Updated Product Stock:")
for product in products:
    product.show()
```

### 21. Project 5: Employee Payroll System


```python
from abc import ABC, abstractmethod

class Employee(ABC):
    def __init__(self, employee_id, name):
        self.employee_id = employee_id
        self.name = name

    @abstractmethod
    def calculate_salary(self):
        pass

    def show_salary(self):
        print(f"ID: {self.employee_id}")
        print(f"Name: {self.name}")
        print(f"Salary: {self.calculate_salary()}")
        print("-" * 40)


class FullTimeEmployee(Employee):
    def __init__(self, employee_id, name, monthly_salary, bonus=0):
        super().__init__(employee_id, name)
        self.monthly_salary = monthly_salary
        self.bonus = bonus

    def calculate_salary(self):
        return self.monthly_salary + self.bonus


class PartTimeEmployee(Employee):
    def __init__(self, employee_id, name, hourly_rate, hours_worked):
        super().__init__(employee_id, name)
        self.hourly_rate = hourly_rate
        self.hours_worked = hours_worked

    def calculate_salary(self):
        return self.hourly_rate * self.hours_worked


class CommissionEmployee(Employee):
    def __init__(self, employee_id, name, base_salary, sales_amount, commission_rate):
        super().__init__(employee_id, name)
        self.base_salary = base_salary
        self.sales_amount = sales_amount
        self.commission_rate = commission_rate

    def calculate_salary(self):
        commission = self.sales_amount * self.commission_rate / 100
        return self.base_salary + commission


class PayrollSystem:
    def __init__(self):
        self.employees = []

    def add_employee(self, employee):
        self.employees.append(employee)

    def process_payroll(self):
        print("Payroll Report")
        print("=" * 40)
        total_salary = 0

        for employee in self.employees:
            employee.show_salary()
            total_salary += employee.calculate_salary()

        print("Total Salary Cost:", total_salary)


payroll = PayrollSystem()

payroll.add_employee(FullTimeEmployee("E-01", "Smith", 50000, 5000))
payroll.add_employee(PartTimeEmployee("E-02", "Nila", 500, 80))
payroll.add_employee(CommissionEmployee("E-03", "Emma", 25000, 200000, 5))

payroll.process_payroll()
```

```python
class Vehicle:
    def __init__(self, vehicle_id, brand, model, rent_per_day):
        self.vehicle_id = vehicle_id
        self.brand = brand
        self.model = model
        self.rent_per_day = rent_per_day
        self.is_available = True

    def calculate_rent(self, days):
        return self.rent_per_day * days

    def rent_vehicle(self):
        if self.is_available:
            self.is_available = False
            return True
        return False

    def return_vehicle(self):
        self.is_available = True

    def show(self):
        status = "Available" if self.is_available else "Rented"
        print(f"{self.vehicle_id} - {self.brand} {self.model} - {self.rent_per_day}/day - {status}")


class Car(Vehicle):
    def __init__(self, vehicle_id, brand, model, rent_per_day, seats):
        super().__init__(vehicle_id, brand, model, rent_per_day)
        self.seats = seats

    def calculate_rent(self, days):
        base_rent = super().calculate_rent(days)
        service_charge = 500
        return base_rent + service_charge


class Bike(Vehicle):
    def __init__(self, vehicle_id, brand, model, rent_per_day, engine_cc):
        super().__init__(vehicle_id, brand, model, rent_per_day)
        self.engine_cc = engine_cc

    def calculate_rent(self, days):
        return super().calculate_rent(days)


class RentalSystem:
    def __init__(self):
        self.vehicles = []

    def add_vehicle(self, vehicle):
        self.vehicles.append(vehicle)

    def find_vehicle(self, vehicle_id):
        for vehicle in self.vehicles:
            if vehicle.vehicle_id == vehicle_id:
                return vehicle
        return None

    def show_available_vehicles(self):
        print("Available Vehicles:")
        for vehicle in self.vehicles:
            if vehicle.is_available:
                vehicle.show()

    def rent_vehicle(self, vehicle_id, days):
        vehicle = self.find_vehicle(vehicle_id)

        if not vehicle:
            print("Vehicle not found.")
            return

        if vehicle.rent_vehicle():
            print("Vehicle rented successfully.")
            print("Total rent:", vehicle.calculate_rent(days))
        else:
            print("Vehicle is already rented.")

    def return_vehicle(self, vehicle_id):
        vehicle = self.find_vehicle(vehicle_id)

        if not vehicle:
            print("Vehicle not found.")
            return

        vehicle.return_vehicle()
        print("Vehicle returned successfully.")


system = RentalSystem()

system.add_vehicle(Car("C-01", "Toyota", "Axio", 3000, 5))
system.add_vehicle(Car("C-02", "Honda", "Civic", 4000, 5))
system.add_vehicle(Bike("B-01", "Yamaha", "FZ", 1200, 150))

system.show_available_vehicles()

system.rent_vehicle("C-01", 3)
system.show_available_vehicles()

system.return_vehicle("C-01")
system.show_available_vehicles()
```

### 23. Project 7: Hospital Appointment System


```python
class Doctor:
    def __init__(self, doctor_id, name, specialty, fee):
        self.doctor_id = doctor_id
        self.name = name
        self.specialty = specialty
        self.fee = fee

    def show(self):
        print(f"{self.doctor_id} | Dr. {self.name} | {self.specialty} | Fee: {self.fee}")


class Patient:
    def __init__(self, patient_id, name, age, phone):
        self.patient_id = patient_id
        self.name = name
        self.age = age
        self.phone = phone

    def show(self):
        print(f"{self.patient_id} | {self.name} | Age: {self.age} | Phone: {self.phone}")


class Appointment:
    next_appointment_id = 1

    def __init__(self, patient, doctor, date, time):
        self.appointment_id = Appointment.next_appointment_id
        Appointment.next_appointment_id += 1

        self.patient = patient
        self.doctor = doctor
        self.date = date
        self.time = time
        self.status = "Scheduled"

    def cancel(self):
        self.status = "Cancelled"

    def show(self):
        print("Appointment ID:", self.appointment_id)
        print("Patient:", self.patient.name)
        print("Doctor:", self.doctor.name)
        print("Specialty:", self.doctor.specialty)
        print("Date:", self.date)
        print("Time:", self.time)
        print("Fee:", self.doctor.fee)
        print("Status:", self.status)
        print("-" * 40)


class HospitalSystem:
    def __init__(self):
        self.doctors = []
        self.patients = []
        self.appointments = []

    def add_doctor(self, doctor):
        self.doctors.append(doctor)

    def add_patient(self, patient):
        self.patients.append(patient)

    def find_doctor(self, doctor_id):
        for doctor in self.doctors:
            if doctor.doctor_id == doctor_id:
                return doctor
        return None

    def find_patient(self, patient_id):
        for patient in self.patients:
            if patient.patient_id == patient_id:
                return patient
        return None

    def create_appointment(self, patient_id, doctor_id, date, time):
        patient = self.find_patient(patient_id)
        doctor = self.find_doctor(doctor_id)

        if not patient:
            print("Patient not found.")
            return

        if not doctor:
            print("Doctor not found.")
            return

        appointment = Appointment(patient, doctor, date, time)
        self.appointments.append(appointment)
        print("Appointment created successfully.")

    def cancel_appointment(self, appointment_id):
        for appointment in self.appointments:
            if appointment.appointment_id == appointment_id:
                appointment.cancel()
                print("Appointment cancelled.")
                return
        print("Appointment not found.")

    def show_all_appointments(self):
        if not self.appointments:
            print("No appointments found.")
            return

        for appointment in self.appointments:
            appointment.show()


hospital = HospitalSystem()

hospital.add_doctor(Doctor("D-01", "Hasan", "Cardiology", 1000))
hospital.add_doctor(Doctor("D-02", "Mim", "Neurology", 1200))

hospital.add_patient(Patient("P-01", "Smith", 30, "01700000"))
hospital.add_patient(Patient("P-02", "Nila", 25, "01800000"))

hospital.create_appointment("P-01", "D-01", "2026-06-20", "10:30 AM")
hospital.create_appointment("P-02", "D-02", "2026-06-21", "12:00 PM")

hospital.show_all_appointments()

hospital.cancel_appointment(1)
hospital.show_all_appointments()
```

### 24. Project 8: Quiz Game 


```python
class Question:
    def __init__(self, question_text, options, correct_option):
        self.question_text = question_text
        self.options = options
        self.correct_option = correct_option

    def check_answer(self, answer):
        return answer.upper() == self.correct_option.upper()

    def show_question(self):
        print(self.question_text)
        for key, value in self.options.items():
            print(f"{key}. {value}")


class Quiz:
    def __init__(self, title):
        self.title = title
        self.questions = []
        self.score = 0

    def add_question(self, question):
        self.questions.append(question)

    def start_quiz(self, answers):
        print("Quiz:", self.title)
        print("=" * 40)

        self.score = 0

        for index, question in enumerate(self.questions):
            question.show_question()

            user_answer = answers[index]

            print("Your answer:", user_answer)

            if question.check_answer(user_answer):
                print("Correct!")
                self.score += 1
            else:
                print("Wrong!")
                print("Correct answer:", question.correct_option)

            print("-" * 40)

        self.show_result()

    def show_result(self):
        total = len(self.questions)
        percentage = (self.score / total) * 100

        print("Final Score:", self.score, "/", total)
        print("Percentage:", percentage)

        if percentage >= 80:
            print("Grade: Excellent")
        elif percentage >= 60:
            print("Grade: Good")
        elif percentage >= 40:
            print("Grade: Passed")
        else:
            print("Grade: Failed")


quiz = Quiz("Python OOP Quiz")

quiz.add_question(Question(
    "Which keyword is used to define a class in Python?",
    {"A": "function", "B": "class", "C": "object", "D": "def"},
    "B"
))

quiz.add_question(Question(
    "Which method is called automatically when an object is created?",
    {"A": "__start__", "B": "__main__", "C": "__init__", "D": "__newclass__"},
    "C"
))

quiz.add_question(Question(
    "Which OOP concept means same method name with different behavior?",
    {"A": "Inheritance", "B": "Polymorphism", "C": "Encapsulation", "D": "Loop"},
    "B"
))

# Demo answers
quiz.start_quiz(["B", "C", "A"])
```

### 26. Project 10: Simple OOP CRUD Operations


```python
class Product:
    def __init__(self, product_id, name, price):
        self.product_id = product_id
        self.name = name
        self.price = price

    def update(self, name=None, price=None):
        if name is not None:
            self.name = name
        if price is not None:
            self.price = price

    def show(self):
        print(f"ID: {self.product_id} | Name: {self.name} | Price: {self.price}")


class ProductManager:
    def __init__(self):
        self.products = {}

    def create_product(self, product_id, name, price):
        if product_id in self.products:
            print("Product already exists.")
            return

        self.products[product_id] = Product(product_id, name, price)
        print("Product created.")

    def read_product(self, product_id):
        product = self.products.get(product_id)

        if product:
            product.show()
        else:
            print("Product not found.")

    def update_product(self, product_id, name=None, price=None):
        product = self.products.get(product_id)

        if product:
            product.update(name, price)
            print("Product updated.")
        else:
            print("Product not found.")

    def delete_product(self, product_id):
        if product_id in self.products:
            del self.products[product_id]
            print("Product deleted.")
        else:
            print("Product not found.")

    def show_all_products(self):
        if not self.products:
            print("No products available.")
            return

        for product in self.products.values():
            product.show()


manager = ProductManager()

manager.create_product("P-01", "Laptop", 80000)
manager.create_product("P-02", "Mouse", 700)
manager.create_product("P-03", "Keyboard", 1200)

manager.show_all_products()

manager.read_product("P-02")

manager.update_product("P-02", price=900)
manager.read_product("P-02")

manager.delete_product("P-01")
manager.show_all_products()
```

### 27. OOP Example: University Course Registration System


```python
from abc import ABC, abstractmethod
from dataclasses import dataclass


# Dataclass
@dataclass
class Address:
    city: str
    area: str
    road: str

    def full_address(self):
        return f"{self.road}, {self.area}, {self.city}"


# Abstraction
class PaymentMethod(ABC):
    @abstractmethod
    def pay(self, amount):
        pass


# Polymorphism
class BkashPayment(PaymentMethod):
    def __init__(self, number):
        self.number = number

    def pay(self, amount):
        return f"Paid {amount} BDT using Bkash"


class CardPayment(PaymentMethod):
    def __init__(self, holder_name):
        self.holder_name = holder_name

    def pay(self, amount):
        return f"Paid {amount} BDT using Card"


# Parent Class
class Person:
    def __init__(self, person_id, name, phone):
        self.person_id = person_id
        self.name = name
        self.phone = phone

    def show_info(self):
        return f"{self.name} | {self.phone}"

    def __str__(self):
        return f"{self.name} ({self.person_id})"


# Inheritance
class Customer(Person):
    total_customers = 0

    def __init__(self, person_id, name, phone, address, wallet):
        super().__init__(person_id, name, phone)
        self.address = address
        self.__wallet = wallet        # Encapsulation
        self.orders = []
        Customer.total_customers += 1

    @property
    def wallet(self):
        return self.__wallet

    def reduce_wallet(self, amount):
        if amount <= self.__wallet:
            self.__wallet -= amount
            return True
        return False

    @classmethod
    def count(cls):
        return cls.total_customers

    def show_info(self):
        return f"Customer: {self.name} | Wallet: {self.__wallet} BDT"


class Rider(Person):
    def __init__(self, person_id, name, phone, vehicle):
        super().__init__(person_id, name, phone)
        self.vehicle = vehicle
        self.deliveries = 0

    def deliver(self, order):
        order.status = "Delivered"
        self.deliveries += 1
        return f"{self.name} delivered Order #{order.order_id}"

    def show_info(self):
        return f"Rider: {self.name} | Vehicle: {self.vehicle} | Deliveries: {self.deliveries}"


class FoodItem:
    def __init__(self, item_id, name, price, category):
        self.item_id = item_id
        self.name = name
        self.price = price
        self.category = category

    @staticmethod
    def valid_price(price):
        return price > 0

    def __eq__(self, other):
        return self.item_id == other.item_id

    def __str__(self):
        return f"{self.name} - {self.price} BDT"


class Restaurant:
    total_restaurants = 0

    def __init__(self, restaurant_id, name, address):
        self.restaurant_id = restaurant_id
        self.name = name
        self.address = address
        self.menu = []
        Restaurant.total_restaurants += 1

    def add_food(self, food):
        if not FoodItem.valid_price(food.price):
            return "Invalid food price."

        if food in self.menu:
            return f"{food.name} already exists."

        self.menu.append(food)
        return f"{food.name} added to menu."

    def find_food(self, item_id):
        for food in self.menu:
            if food.item_id == item_id:
                return food
        return None

    @classmethod
    def count(cls):
        return cls.total_restaurants

    def __len__(self):
        return len(self.menu)

    def __str__(self):
        return f"{self.name} - {self.address.area}"


# Composition
class CartItem:
    def __init__(self, food, quantity):
        self.food = food
        self.quantity = quantity

    def subtotal(self):
        return self.food.price * self.quantity

    def __str__(self):
        return f"{self.food.name} x {self.quantity} = {self.subtotal()} BDT"


class Cart:
    def __init__(self):
        self.items = []

    def add_item(self, food, quantity):
        if food is None:
            return "Food not found."

        if quantity <= 0:
            return "Invalid quantity."

        for item in self.items:
            if item.food == food:
                item.quantity += quantity
                return f"{food.name} quantity updated."

        self.items.append(CartItem(food, quantity))
        return f"{food.name} added to cart."

    def total(self):
        return sum(item.subtotal() for item in self.items)

    def __len__(self):
        return len(self.items)


# Abstraction
class Discount(ABC):
    @abstractmethod
    def apply(self, amount):
        pass


class StudentDiscount(Discount):
    def apply(self, amount):
        return amount * 0.90


class RegularDiscount(Discount):
    def apply(self, amount):
        return amount


class Order:
    next_id = 1001

    def __init__(self, customer, restaurant, cart, discount):
        self.order_id = Order.next_id
        Order.next_id += 1

        self.customer = customer
        self.restaurant = restaurant
        self.cart = cart
        self.discount = discount
        self.status = "Pending"

    def delivery_charge(self):
        return 60

    def final_bill(self):
        discounted_amount = self.discount.apply(self.cart.total())
        return discounted_amount + self.delivery_charge()

    def checkout(self, payment_method):
        bill = self.final_bill()

        if not self.customer.reduce_wallet(bill):
            return "Order failed. Insufficient wallet balance."

        self.customer.orders.append(self)
        self.status = "Confirmed"

        return f"Order #{self.order_id} confirmed. {payment_method.pay(bill)}"

    def __str__(self):
        return f"Order #{self.order_id} - {self.status}"


# Object Creation

customer = Customer(
    "C-101",
    "Smith",
    "01700000000",
    Address("Dhaka", "Dhanmondi", "Road 10"),
    2000
)

rider = Rider(
    "R-501",
    "Emma",
    "01800000000",
    "Bike"
)

restaurant = Restaurant(
    "RES-01",
    "Food Palace",
    Address("Dhaka", "Banani", "Road 2")
)

foods = [
    FoodItem("F-01", "Chicken Burger", 250, "Fast Food"),
    FoodItem("F-02", "Beef Pizza", 650, "Fast Food"),
    FoodItem("F-03", "Fried Rice", 300, "Rice Item"),
    FoodItem("F-04", "Cold Coffee", 150, "Drinks")
]


# Add Food Items

for food in foods:
    restaurant.add_food(food)


# Create Cart

cart = Cart()

cart.add_item(restaurant.find_food("F-01"), 2)
cart.add_item(restaurant.find_food("F-03"), 1)
cart.add_item(restaurant.find_food("F-04"), 1)


# Create Order 
order = Order(
    customer,
    restaurant,
    cart,
    StudentDiscount()
)

# Dependency Injection
checkout_message = order.checkout(
    BkashPayment("0111111111")
)

delivery_message = rider.deliver(order)


print("FOOD DELIVERY SYSTEM - OOP FINAL EXAMPLE")
print("-" * 50)

print("Customer:", customer.name)
print("Address:", customer.address.full_address())
print("Restaurant:", restaurant.name)
print("Rider:", rider.name)

print("\nCart Summary:")
for item in cart.items:
    print("-", item)

print("\nCheckout:")
print(checkout_message)

print("\nDelivery:")
print(delivery_message)

print("\nFinal Summary:")
print("Order ID:", order.order_id)
print("Order Status:", order.status)
print("Food Total:", cart.total(), "BDT")
print("Final Bill:", order.final_bill(), "BDT")
print("Customer Wallet Balance:", customer.wallet, "BDT")
print("Total Customers:", Customer.count())
print("Total Restaurants:", Restaurant.count())
print("Menu Items:", len(restaurant))
print("Cart Items:", len(cart))
print("Order:", order)
```
